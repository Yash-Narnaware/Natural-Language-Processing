```python
tmdb = pd.read_csv('tmdb_5000_movies.csv/tmdb_5000_movies.csv')
```

```python
tmdb.head()
```

```python
# Keep only the required columns
columns_to_keep = ['id', 'title', 'genres', 'keywords', 'overview']
movies_cleaned = tmdb[columns_to_keep].copy()
```

```python
# Drop rows with missing essential fields
movies_cleaned.dropna(subset=['genres', 'keywords', 'overview'], inplace=True)

# Reset index for consistency
movies_cleaned.reset_index(drop=True, inplace=True)
```

```python
movies_cleaned
```

```python
import ast  # Safe parser for stringified Python literals

def extract_names(obj):
    try:
        items = ast.literal_eval(obj)
        return ' '.join([item['name'] for item in items])
    except:
        return ''
```

```python
movies_cleaned
```

```python
movies_cleaned['genres'] = movies_cleaned['genres'].apply(extract_names)
movies_cleaned['keywords'] = movies_cleaned['keywords'].apply(extract_names)
```

```python
# Combine overview, genres, and keywords into one text field
movies_cleaned['tags'] = (
    movies_cleaned['overview'] + ' ' +
    movies_cleaned['genres'] + ' ' +
    movies_cleaned['keywords']
)
```

```python
# Optional: lowercase for consistency
movies_cleaned['tags'] = movies_cleaned['tags'].str.lower()
```

```python
movie_final = movies_cleaned[['id', 'title', 'tags']]
```
